```python
import pandas as pd
import json
from transformers import BertTokenizer
```

```python
df = pd.read_csv('preprocessed_main_df.csv')
df.head()
```

```python
def get_tokens(review_json):
  try:
    decoded_reviews = json.loads(review_json)
    text = " ".join(decoded_reviews)
    tokens = text.split()
    return tokens
  except json.JSONDecodeError:
    return []

df['User review'] = df['User review'].apply(get_tokens)
df.head()
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-uncased')
```

```python
reviews = df['User review'].apply(lambda x: ' '.join(x)).tolist()
tokens = tokenizer(reviews, padding=True, truncation=True, return_tensors="tf")
print(tokens)
#print(reviews[0])
```

```python
df['input_ids'] = tokens['input_ids'].numpy().tolist()
df['attention_mask'] = tokens['attention_mask'].numpy().tolist()
df.head()
```

```python
df.to_csv('Bert_df.csv')
files.download('Bert_df.csv')
```
