```python
import pandas as pd
import json
from sentence_transformers import SentenceTransformer
```

```python
df = pd.read_csv('preprocessed_main_df.csv')
df.head()
```

```python
def get_tokens(review_json):
  try:
    decoded_reviews = json.loads(review_json)
    text = " ".join(decoded_reviews)
    tokens = text.split()
    return tokens
  except json.JSONDecodeError:
    return []

df['User review'] = df['User review'].apply(get_tokens)
df.head()
```

```python
df['User review'] = df['User review'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))
df['User review'] = df['User review'].fillna('')

model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

embeddings = model.encode(df['User review'].tolist())
print(embeddings)
```

```python
df['User review'] = embeddings.tolist()
df.head()
```
